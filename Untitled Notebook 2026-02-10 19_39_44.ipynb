```python
import datetime

class Logger:
    def __init__(self, name=None):
        self.name = name or "Logger"

    def _format(self, level, msg):
        ts = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        return f"{ts} [{self.name}] {level}: {msg}"

    def info(self, msg):
        print(self._format("INFO", msg))

    def warn(self, msg):
        print(self._format("WARN", msg))

    def error(self, msg):
        print(self._format("ERROR", msg))

    def debug(self, msg):
        print(self._format("DEBUG", msg))


if __name__ == "__main__":
    logger = Logger("MyApp")
    logger.info("hellow message")
```
